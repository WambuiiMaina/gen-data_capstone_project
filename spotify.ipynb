```python
def search_show(token, podcast_name):
    url = "https://api.spotify.com/v1/search"
    headers = {"Authorization": f"Bearer {token}"}
    params = {"q": podcast_name, "type": "show", "limit": 1}  # get the best match
    res = requests.get(url, headers=headers, params=params)
    items = res.json().get("shows", {}).get("items", [])
    return items[0] if items else None
```

```python
def get_episodes(token, show_id):
    url = f"https://api.spotify.com/v1/shows/{show_id}/episodes"
    headers = {"Authorization": f"Bearer {token}"}
    episodes = []

    while url:
        res = requests.get(url, headers=headers).json()
        episodes.extend(res.get("items", []))
        url = res.get("next")  # pagination
    return episodes
```

```python
import pandas as pd
```

```python
all_data = []

for podcast in podcast_names:
    show = search_show(token, podcast)
    if not show:
        print(f"Show not found: {podcast}")
        continue

    show_id = show["id"]
    episodes = get_episodes(token, show_id)

    if not episodes:
        print(f"No episodes found: {podcast}")
        continue

# filter valid responses
valid_episodes = [ep for ep in episodes if ep and ep.get("duration_ms")]

if not valid_episodes:
    print("No valid episode durations found")
    
#Calculate average duration
avg_duration = sum(ep["duration_ms"] for ep in valid_episodes) / 60000 / len(valid_episodes)

# Calculate posting frequency per month
dates = [datetime.strptime(ep["release_date"], "%Y-%m-%d") for ep in episodes if "release_date" in ep]
dates.sort()
if len(dates) > 1:
        months_active = (dates[-1] - dates[0]).days / 30
        posting_freq = len(episodes) / months_active if months_active > 0 else len(episodes)
else:
        posting_freq = len(episodes)

all_data.append({
        "podcast_name": show["name"],
        "show_id": show["id"],
        "publisher": show.get("publisher", "N/A"),
        "total_episodes": show["total_episodes"],
        "avg_duration_min": round(avg_duration, 1),
        "posting_frequency_per_month": round(posting_freq, 2),
        "description": show.get("description", "N/A")
    })

# Convert to DataFrame
df = pd.DataFrame(all_data)

# Save to Excel
df.to_excel("kenya_podcasts.xlsx", index=False)
print("✅ Dataset saved to kenya_podcasts.xlsx")
print(df)
```
