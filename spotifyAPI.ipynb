```python
import spotipy
import pandas as pd
import time
```

```python
from spotipy.oauth2 import SpotifyClientCredentials

client_id = "a8c9b726e6874f679d991c3362dec91e"
client_secret = "dd0b4f683e0243058a19a6353d686c83"

auth_manager = SpotifyClientCredentials(
    client_id=client_id,
    client_secret=client_secret
)

sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
results_per_page = 10  # you can set up to 50
total_results_to_fetch = 50  # e.g., get 50 shows
show_names = []
show_ids = []

for offset in range(0, total_results_to_fetch, results_per_page):
    results = sp.search(q="Kenyan podcast", type="show", limit=results_per_page, offset=offset)
    for show in results['shows']['items']:
        show_names.append(show['name'])
        show_ids.append(show['id'])

# Print all shows collected
for show in zip(show_names, show_ids):
    print(show[0], '-', show[1])

print(f"Total shows collected: {len(show_ids)}")
```

```python
#sample show id for testing
show_id = '2VcFFAEGIsL3M9BR0M9FAk'

# Get show details
show_details = sp.show(show_id)

print("Show Name:", show_details.get('name'))
print("Publisher:", show_details.get('publisher', 'Unknown'))
print("Total Episodes:", show_details.get('total_episodes'))

# Get 10 episodes
episodes = sp.show_episodes(show_id, limit=10)

for ep in episodes.get('items', []):
    print("Episode Name:", ep.get('name'))
    print("Release Date:", ep.get('release_date'))
    print("Duration (ms):", ep.get('duration_ms'))
    print("Explicit:", ep.get('explicit'))
    print("-" * 40)
```

```python
# getting basic info about the podcasts(shows)
show_data = []
for show_id in show_ids:
    try:
        show_info = sp.show(show_id)
        show_data.append({
            'name': show_info['name'],
            'total_episodes': show_info['total_episodes'],
            'description': show_info['description']
        })
    except Exception as e:
        print(f"Error fetching data for show ID {show_id}: {e}")
    time.sleep(1)  # to avoid hitting rate limits
```

```python
# getting the episode data for the podcasts
episode_data = []
for show_id in show_ids:
    try:
        episodes = sp.show_episodes(show_id, limit=50)
        for episode in episodes['items']:
            episode_data.append({
                'show_id': show_id,
                'episode_name': episode['name'],
                'release_date': episode['release_date'],
                'duration_ms': episode['duration_ms']
            })
    except Exception as e:
        print(f"Error fetching episodes for show ID {show_id}: {e}")
    time.sleep(1)  # to avoid hitting rate limits
```

```python
#saving the data to excel
import pandas as pd

# Convert to DataFrame
df_shows = pd.DataFrame(show_data)
df_episodes = pd.DataFrame(episode_data)

# Save DataFrame to Excel
df_shows.to_excel('shows_data.xlsx', index=False)
df_episodes.to_excel('episodes_data.xlsx', index=False)
print("Saved as shows_data.xlsx and episodes_data.xlsx")
```
